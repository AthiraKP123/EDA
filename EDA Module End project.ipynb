#### Q1. How many are there in each Team and the percentage splitting with respect to the total employees.

```python
df1=pd.DataFrame({"Counts":df.Team.value_counts()}).reset_index()
df1['Short'] = df1['Team'].apply(lambda x: ''.join(word[0] for word in x.split()))
df1
```

### Data story
Here is the total number of employees in each team and created an abbreviation for each team name so that it will be easier for data visualisation. Then craeted a new dataframe called "df1" to store only the required data such as Team name,Abbreviation, and the number of employees un each team.

```python
plt.figure(figsize=(15, 10))
plt.bar(df1["Short"],df1["Counts"],color=plt.cm.tab20c.colors)
plt.title("No.of employees in each team")
plt.xlabel("Team")
plt.ylabel("No.of Employees")
plt.show()
```

Here is the visual representation of total employees in each team using bar chart. x-axis is used to represent the team name and the y-axis represents the total no of employees

```python
df1["Percentage"] = (df1["Counts"]/df1["Counts"].sum())*100
df1
```

### Data story
created a new column named "Percentages" which shows the percentage split of employees.
It is calculated by dividing the count of employees any one team with the total number of employees and then multiplied by 100 to get the percentage.

```python
plt.figure(figsize=(11,6))
plt.plot(df1['Short'], df1['Percentage'], marker='o')
plt.title("Percentage split of employees")
plt.xlabel("Short")
plt.ylabel("Percentage")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,6))
plt.pie(df1['Percentage'],labels=df1['Short'])
plt.title("Percentage split of employees")
plt.show()
```

### Data Story
From analysing the above line graph , it's clear that the team "New Orleans Pelicans" has more percentage split compared to other teams , Also there is a identical pattern for some teams starting from "BN(Brooklyn Nets)" to "DM(Dallas Mavericks)".

#### Q4. Find out under which team and position, spending in terms of salary is high.

```python
grouped_df = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
sorted_df = grouped_df.sort_values(by='Salary', ascending=False)
sorted_df['Abbreviation'] = sorted_df['Team'].apply(lambda x: ''.join(word[0] for word in x.split()))
sorted_df
```

### Data story
Here we have created a data frame with the columns Team, Position, Salary and Abbreviation. The column "Salary" displays the sum of salaries of all employees of a particular position within a team. The data is displayed in descending order with respect to the values in "Salary".

```python
plt.figure(figsize=(20, 30))
plt.barh(y=sorted_df['Abbreviation'] + ' - ' + sorted_df['Position'],width=sorted_df['Salary'],color=plt.cm.tab20c.colors)
plt.xlabel('Salary')
plt.ylabel('Team-Position')
plt.title('Position vs Salary Graph')
plt.show()
```

### Data story
here is the Bar graph representation of Position within a team and the salary. Y axis represents each position in the team and X axis represents salary.
